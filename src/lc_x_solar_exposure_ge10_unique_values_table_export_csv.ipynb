```python
import pandas as pd

# Load the unique values table
df = pd.read_csv(r"C:\Users\saral\SL_Website\LandCoverMapping\03_QGIS\3_gpkg_csv\lc_x_solar_exposure_ge10_unique_values_table.csv")

# Keep only main LC classes
df = df[df["lc_class"].isin([10, 20, 30, 40])]

# Pivot: rows = LC, columns = solar class, values = % of LC
pivot = (
    df.pivot_table(
        index="lc_class",
        columns="solar",
        values="lc_share",
        aggfunc="sum"
    )
    .fillna(0)
)

# Optional: rename columns for clarity
pivot.columns = [f"solar_{int(c)}" for c in pivot.columns]

# Sanity check (each row should sum to ~1.0)
pivot["row_sum"] = pivot.sum(axis=1)
print(pivot)

# Export to CSV
pivot.drop(columns="row_sum").to_csv(
    r"C:\Users\saral\SL_Website\LandCoverMapping\04_Python\output\lc_x_solar_exposure_ge10_LC_normalized.csv",
    float_format="%.4f"
)
```
